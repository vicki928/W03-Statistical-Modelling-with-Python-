# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import pandas as pd
import requests
```

```python


url = 'http://api.citybik.es/v2/networks/mobibikes'
city_bikes = requests.get(url).json()
city_bikes




```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
#finding the stations in that city
stations = city_bikes['network']['stations']
stations
#finding latitude and longitude and free bikes for each stataion
latitude = stations[0]['latitude']
latitude
latitude_list = []
#finding longitude for each statation
longitude = stations[0]['longitude']
longitude
longitude_list = []
#find free bikes for each station
free_bikes = stations[0]['free_bikes']
free_bikes
free_bikes_list = []
#find station id
stationsid = stations[0]['id']
stationsid
stationsid_list = []


#get the list of data for latitude,longitude,free_bikes
for i in range(len(stations)):
    latitude_list.append(stations[i]['latitude'])
    longitude_list.append(stations[i]['longitude'])
    free_bikes_list.append(stations[i]['free_bikes'])
    stationsid_list.append(stations[i]['id'])
    
```

Put your parsed results into a DataFrame.

```python
s1=pd.Series(latitude_list)
s2=pd.Series(longitude_list)
s3=pd.Series(free_bikes_list)
s4=pd.Series(stationsid_list)

df_bike1=pd.DataFrame(latitude_list,columns= ["latitude"])
df_bike2=pd.DataFrame(longitude_list,columns= ["longitude"])
df_bike3=pd.DataFrame( free_bikes_list,columns= ["free_bikes"])
df_bike4=pd.DataFrame( stationsid_list,columns= ["stationsid"])
df=pd.concat([df_bike1,df_bike2,df_bike3,df_bike4],axis=1)
df.head()
```
